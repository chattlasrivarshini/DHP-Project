Retail Sales Analysis

```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\karth\OneDrive\Desktop\DHP Lab\Class\train.csv")
```

1.Find the total sales for each city.

```python
city_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False)
print(city_sales)
```

2.	Identify the top 5 products with the highest sales.

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5)
print(top_products)
```

3.	Calculate total sales for each category.

```python
category_sales = df.groupby('Category')['Sales'].sum()
print(category_sales)
```

4.	Find the customer who generated the highest total sales.

```python
top_customer = df.groupby('Customer Name')['Sales'].sum().idxmax()
top_customer_sales = df.groupby('Customer Name')['Sales'].sum().max()

print(top_customer, top_customer_sales)
```

5.	Determine the city with the highest total sales.

```python
top_city = df.groupby('City')['Sales'].sum().idxmax()
top_city_sales = df.groupby('City')['Sales'].sum().max()

print(top_city, top_city_sales)
```

6.	Find the total sales for each sub-category.

```python
subcat_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
print(subcat_sales)
```

7.	Calculate the monthly sales trend using Order Date.

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month

monthly_sales = df.groupby(['Year', 'Month'])['Sales'].sum()

print(monthly_sales)
```

8.	Identify orders with sales greater than 1000.

```python
high_sales_orders = df[df['Sales'] > 1000]
print(high_sales_orders)
```

9.	Calculate the average sales per customer.

```python
avg_sales_customer = df.groupby('Customer Name')['Sales'].mean()
print(avg_sales_customer)
```

10.	Identify the least-selling product (based on total sales)

```python
least_product = df.groupby('Product Name')['Sales'].sum().idxmin()
least_product_sales = df.groupby('Product Name')['Sales'].sum().min()

print(least_product, least_product_sales)
```
